```python
# Importar as bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import os

# Mapear o caminho do arquivo
caminhoArquivo = './arquivos/company_sales_data.csv'

# Alocar os dados em um dataframe
df = pd.read_csv(caminhoArquivo)

# Visulizar o cabeçalho do Dataframe
df.head()
```

```python
# Selecionar os dados que vão ser apresentados no gráfico
listaLucro = df['total_profit'].to_list()
listaMeses = df['month_number'].to_list()

# Criar o gráfico utilizandos os dados listados anteriormente
plt.plot(listaMeses, listaLucro, label = 'Lucro mensal do ano anterior')

plt.xlabel('Número do mês') # Título do eixo X
plt.ylabel('Lucro') # Título do eixo Y
plt.title('Lucro mensal da empresa') # Título do gráfico
plt.legend() # Legenda do gráfico
```
