### Trabalhando com Pandas SERIES
Uma série em Pandas é como uma coluna de uma tabela. é uma matriz unidimensional que contém dados de qualquer tipo. podemos usar a função pd.serie() para criar uma série.

```python
# importando a biblioteca
import pandas as pd

# criando um pandas series a partir de uma lista
idades = pd.Series([10,20,30,40,50,60])

print(type(idades))
```

## Nomeando as colunas do DataFrame
podemos usar o df.columns para modificar todos os nomes das colunas:


```python
#Modificar os nomes
# as colunas devem estar na mesma ordem dos nomes
# A quantidade de nomes deve ser igual a quantidade de colunas

df.columns =[
    'idade','job','estado civil','educação','default','balance',
    'housing','empréstimo','contato','dia','mês',
    'duração','campanha','pdays','previous','poutcome'
]
df.head
```

### Método df.rename()

```python
df = df.rename(columns={
    'idade':'IDADE',
    'job':'Profissões'
})
df.head(1)
```
