## importar a biblioteca, carregar o arquivo e alocá-lo em um dataFrame

```python
import pandas as pd
#mapear caminho do arquivo
caminho = "C:\\Users\\FIC\\Documents\\luana-cristina-python\\world_happiness_report_2015.csv"

# alocar os dados em um DataFrame
df=pd.read_csv(caminho)
# verificando o cabeçalho
df.head()
#verificando o final do DataFrame
df.tail()
#informações da coluna
df.info()
```
