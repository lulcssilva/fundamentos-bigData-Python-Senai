```python
#importar as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

#mapear o caminho do arquivo
caminhoArquivo = '../luana-cristina-python/company_sales_data.csv'

#alocar os dados em um dataframe
df = pd.read_csv(caminhoArquivo)

#visualizar o cabeçalho do DataFrame
df.head()
```

```python
# selecionar os dados que vão ser apresentados no gráfico
listaLucro = df['total_profit'].to_list()
listaMeses = df['month_number'].to_list()

# Criar o gráfico utilizando os dados listados anteriormente
plt.plot(listaMeses,listaLucro, label = 'lucro mensal do ano anterior') 

plt.xlabel('Número do mês') #Titulo do eixo X
plt.ylabel('Lucro') #Titulo do eixo Y
plt.title('Lucro mensal da empresa') #Título do gráfico
plt.legend() #Legenda do gráfico
```

### Ex 2 

```python
#selecionar os dados que serão apresentados no gráfico
totalVendido = df['total_units'].to_list()
listaMeses = df['month_number'].to_list()

#Gráfico dos dados listados
plt.plot(
    listaMeses,
    totalVendido,
    label = 'Total vendido por mês',
    c= 'red',
    marker = 'o',
    markerfacecolor='k',
    ls = '--',
    linewidth=3
    )
plt.xlabel('Meses')
plt.ylabel('Vendas em dolares')
plt.title('Vendas por mês do ano anterior')
plt.yticks([10000,20000,30000,40000,50000])
plt.xticks(listaMeses)

plt.legend(loc = 'lower right')
```

### Múltiplas curvas no mesmo mecanimos de gráfico


```python
#importar as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

x= np.linspace(0,2 * np.pi, 100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y, label= 'função seno', color= '#853df2')
plt.plot(x,z, label = 'função cosseno', color = '#f53EE4')

#adicionar legenda ao grafico
plt.legend()

#adicionar um titulo ao gráfico
plt.title('função Trigonométrica')

#adicionar um título ao gráfico
plt.xlabel('valores de X')
plt.ylabel('valores de Y')

#Salvar o gráfico como imagem
plt.savefig('./graficos/fun.png')
```

### Exemplo 1 - Gráfico de pizza

```python
# Importar a biblioteca
import matplotlib.pyplot as plt

#Dados para plotar no gráfico
languages = ['java','python','PHP','javaScript','C#','C++']
popularidade = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

#explosão das fatias
explosao =[0.1,0,0,0,0,0]

#definindo cores
cores = ['#DBA99C','#FFBADB', '#7DB9DB','#7DDBB3','#C48CE0','#10D5E1']

#Gerar o gráfico
#label = rótulo de categotia | autopct =rótulos numericos
#%0.1f%% formatando os valores dos rótulos númericos com 1 casa após a virgula
#shadow = sombra

plt.pie(popularidade, labels = languages, autopct='%1.1f%%', shadow=True, explode=explosao,
        colors= cores )

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#mapear o caminho e arquivo
caminhoArquivo = './sea-level.csv'

#alocar os dados em um dataFrame
dados = pd.read_csv(caminhoArquivo)

# fatiar o dataFrame(selecionar a partir da linha 5 e todas as colunas)
dados = dados.loc[6:,:]
dados.columns =[
                'Year',
                'CSIRO - Adjusted sea level (inches)',
                'CSIRO - Lower error bound (inches)',
                'CSIRO - Upper error bound (inches)',
                'NOAA - Adjusted sea level (inches)'
             ]

#Converter os dados em númericos
dados['Year']  = dados['Year'].astype(int)
dados['CSIRO - Adjusted sea level (inches)'] = dados['CSIRO - Adjusted sea level (inches)'].astype(float)
dados['CSIRO - Lower error bound (inches)'] = dados['CSIRO - Lower error bound (inches)'].astype(float)
dados['CSIRO - Upper error bound (inches)'] = dados['CSIRO - Upper error bound (inches)'].astype(float)
dados['NOAA - Adjusted sea level (inches)'] = dados['NOAA - Adjusted sea level (inches)'].astype(float)

dados.info()
```

```python
anos = dados ['Year'].to_list()
nivel = dados['CSIRO - Adjusted sea level (inches)'].to_list()

sns.scatterplot(x=anos, y= nivel)
```

```python
#importar as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

#mapear caminho do arquivo
caminhoArquivo = './company_sales_data.csv'

#alocar os dados em um dataframe
df = pd.read_csv(caminhoArquivo)

#gerar listas
listaMeses = df['month_number'].to_list()
vendasCremeDental = df['toothpaste'].to_list()

#criar gráfico
sns.scatterplot(x = listaMeses, y = vendasCremeDental)

#adicionar grade ao grafico
plt.grid(True, linewidth = 0.5, linestyle = '--', c='#FFBADB')

#adicionar título no gráfico
plt.title('Vendas de creme dental')
plt.xlabel('Meses')
plt.ylabel('Unidades vendidas')
```
